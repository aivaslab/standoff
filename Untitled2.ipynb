```python
import matplotlib.pyplot as plt
import numpy as np

# Define particles as (name, mass, is_fermion)
particles = [("Electron", 0.511, True), ("Proton", 938.3, True), ("Photon", 0.0, False), ("W Boson", 80.379e3, False)]

fig, ax = plt.subplots()

# Constants for visualization
energy_scale = 1000  # Scale factor for visual representation

for i, (name, mass, is_fermion) in enumerate(particles):
    radius = energy_scale / (mass if mass != 0 else 1)  # Avoid division by zero for photons
    circle = plt.Circle((i*2.5, 0), radius, color='blue' if is_fermion else 'red', fill=False, linestyle='--' if is_fermion else '-')
    ax.add_artist(circle)
    plt.text(i*2.5, 0, name, horizontalalignment='center', verticalalignment='center')

ax.set_xlim(-2, 10)
ax.set_ylim(-5, 5)
ax.set_aspect('equal', adjustable='datalim')
plt.axis('off')  # Turn off the axis
plt.show()
```

```python
# Define particles as (name, mass (GeV), is_fermion)
particles = [("Electron", 0.000511, True), ("Proton", 0.938, True), 
             ("Photon", 0.0001, False), ("W+ Boson", 80.379, False), 
             ("W- Boson", 80.379, False), ("Gluon", 0.0001, False)]

fig, ax = plt.subplots()

# Constants for visualization
log_base = 10  # Base for logarithmic scaling

# Starting point for the first ring
current_radius = 0

for name, mass, is_fermion in particles:
    # Calculate radius using log scale
    energy = mass * 938.2720813  # Convert mass (in GeV) to energy (in MeV)
    radius = 0.1* np.log(energy) / np.log(log_base)
    
    # Draw the circle
    circle = plt.Circle((1, 0), current_radius + radius, color='blue' if is_fermion else 'red', fill=False, linestyle='--' if is_fermion else '-')
    ax.add_artist(circle)
    
    # Place text
    plt.text(1, current_radius + radius/2, name, horizontalalignment='center')
    
    # Update current_radius to the outer edge of the current ring
    current_radius = 0

# Handling gluons (8 types)
for i in range(8):
    angle = 2 * np.pi * i / 8
    x, y = (current_radius + radius) * np.cos(angle), (current_radius + radius) * np.sin(angle)
    circle = plt.Circle((x + 1, y), radius, color='green', fill=False)
    ax.add_artist(circle)
    plt.text(x + 1, y, 'Gluon', horizontalalignment='center', verticalalignment='center')

ax.set_aspect('equal')

xx = 1
ax.set_xlim(-xx, xx+1)
ax.set_ylim(-xx, xx)
plt.axis('off')  # Turn off the axis
plt.show()
```
